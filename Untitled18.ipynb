```python
# script_procesar_estudiantes.py
"""
Script para procesar promedios de estudiantes:
- rellena valores faltantes
- calcula promedio semestral
- clasifica estado (Aprobado / Reprobado)
- imprime y grafica resultados
"""

import pandas as pd
import matplotlib.pyplot as plt

# CONFIGURACIÓN / CONSTANTES
MISSING_FILL = 5.0            # valor para rellenar parciales faltantes
UMBRAL_APROBADO = 7.0         # >= 7 → Aprobado
PARCIALES = ['programacion', 'calculo', 'ingles']

# DATOS DE ENTRADA
datos = {
    'Alumno': ['Ana', 'Luis', 'Pedro', 'Marta', 'Julia', 'Carlos'],
    'programacion': [8, 7, 8, 9, 10, 5],
    'calculo': [9, 8, 7, 8, 9, 6],
    'ingles': [10, 8, 8, 7, 8, 8]
}

def procesar_promedios(Promedio_semestral: pd.DataFrame) -> pd.DataFrame:
    """
    Rellena NaN, calcula promedio final y asigna una etiqueta Estado_de_calificacion.
    Devuelve un DataFrame con columnas nuevas: 'Promedio_final' y 'Estado_de_calificacion'.
    """
    # Asegurar columnas esperadas
    for col in PARCIALES:
        if col not in Promedio_semestral.columns:
            Promedio_semestral[col] = pd.NA

    # Rellenar NaNs en parciales con valor configurable
    Promedio_semestral[PARCIALES] = Promedio_semestral[PARCIALES].fillna(MISSING_FILL).astype(float)

    # Calcular promedio semestral (vectorizado)
    Promedio_semestral['Promedio_final'] = Promedio_semestral[PARCIALES].mean(axis=1)

    # Asignar estado (solo Aprobado o Reprobado)
    def etiqueta_estado(prom):
        if prom >= UMBRAL_APROBADO:
            return "Aprobado"
        return "Reprobado"

    Promedio_semestral['Estado_de_calificacion'] = Promedio_semestral['Promedio_final'].apply(etiqueta_estado)
    return Promedio_semestral

def imprimir_resumen(Promedio_semestral: pd.DataFrame):
    print("Datos procesados:")
    print(Promedio_semestral[['Alumno'] + PARCIALES + ['Promedio_final', 'Estado_de_calificacion']].to_string(index=False))
    print("\n--- LISTA DE REPROBADOS ---")
    reprobados = Promedio_semestral.loc[Promedio_semestral['Estado_de_calificacion'] == "Reprobado", 'Alumno'].tolist()
    if reprobados:
        for a in reprobados:
            print(a)
    else:
        print("Ninguno")

def graficar(Promedio_semestral: pd.DataFrame):
    # Histograma del promedio final
    plt.figure(figsize=(8, 4))
    plt.hist(Promedio_semestral['Promedio_final'], bins=8)
    plt.title('Histograma de promedio final')
    plt.xlabel('Promedio final')
    plt.ylabel('Conteo')
    plt.tight_layout()
    plt.show()

    # Barras horizontales por alumno (ordenado)
    Promedio_semestral_sorted = Promedio_semestral.sort_values('Promedio_final', ascending=True)
    plt.figure(figsize=(8, 4))
    plt.barh(Promedio_semestral_sorted['Alumno'], Promedio_semestral_sorted['Promedio_final'])
    plt.xlabel('Promedio final')
    plt.title('Promedio por alumno')
    plt.xlim(0, 10)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    Promedio_semestral = pd.DataFrame(datos)
    Promedio_semestral = procesar_promedios(Promedio_semestral)
    imprimir_resumen(Promedio_semestral)
    graficar(Promedio_semestral)
```
